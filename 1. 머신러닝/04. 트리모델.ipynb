## 데이터 불러오기

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 아이리스 데이터 불러오기 
iris_data = load_iris()

# 훈련, 테스트 데이터 셋 만들기 
X_train, X_test, y_train, y_test = train_test_split(
    iris_data.data, # 데이터 x
    iris_data.target, #  데이터 y
    test_size = 0.2, # 테스트 데이터의 비율
    random_state=11 # 랜덤 규칙 설정 
)
```

## 결정트리모델 생성

```python
from sklearn.tree import DecisionTreeClassifier

# 모델 생성 
dt_clf = DecisionTreeClassifier()

# 모델학습 
dt_clf.fit(X_train, y_train)
```

## Graphviz 시각화
- 트리 모델이 어떻게 훈련을 했는지 시각화

```python
# 시각화를 위한 graphviz
from sklearn.tree import export_graphviz

export_graphviz(
    dt_clf, # 모델 
    out_file ='tree.dot', # 시각화 결과물을 저장할 파일의 경로
    class_names=iris_data.target_names, # 클래스(y)들의 이름
    feature_names= iris_data.feature_names, # 각 특성(x)의 이름 
    impurity=True,
    filled = True 
)
```

```python
import graphviz

with open('tree.dot') as f:
  dot_graph = f.read()

graphviz.Source(dot_graph)
```
